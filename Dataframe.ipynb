A customer wants a product P at the rate of 5000 units per day. This product P have three operations A, B & C, which requires man minutes 5, 8, 6 respectively. The company works for two shifts per day and working time per employee is 420 minutes. Normal efficiency of the operations of the company is 85%.

Demand of P per day (n) = 5000
Man minutes per product (t) = 5 + 8 + 6
                                              = 19 man minutes
Working time per employee (w) = 420 minutes
No of shifts per day (s) = 2
Manpower required per shift = ((5000 * 19)/420)/2
                                               = 113.09
Manpower required per shift considering 85% efficiency = 113.09/.85
                                                                                            = 133.04
                                                                                            = 134 nos
