Intercept = ymean ‚Äì slope* xmean

```python
import numpy as np
```

```python
X = np.array((1,2,3,4,5))
Y = np.array((1,2,3,4,5))
```

```python
Xi = X 
Yi = Y
    
#Total number of data points:
n = len(Xi)
    

#X bar:
Xi_mean = Xi.mean()

#Y bar:
Yi_mean = Yi.mean()
    

#Sum of X:
S_Xi = (Xi).sum()
    

#Sum of Y:
S_Yi = (Yi).sum()
    

#Sum of (X*Y) multiplied by n:
S_XiYi = ((Xi*Yi).sum())*n
    
#Sum of X*Sum of Y:
S_Xi_S_Yi = S_Xi*S_Yi
    
#Sum of (X*X) multiplied by n:
S_XiXi = ((Xi*Xi).sum())*n
    
#Square of sum of X:
S_Xi_Square = S_Xi*S_Xi
    
#Slope:
dummy_slope = (S_XiYi- S_Xi_S_Yi) / (S_XiXi-S_Xi_Square)
    
#Intercept:
dummy_intercept = Yi_mean - dummy_slope * Xi_mean
    
```

```python
#Check the function with dummy data:
y= (dummy_slope * X)  +  dummy_intercept
print ("Slope : " , dummy_slope)
print ("Intercept :",dummy_intercept)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
plt.plot(X,y,color="red")
plt.show()
```
